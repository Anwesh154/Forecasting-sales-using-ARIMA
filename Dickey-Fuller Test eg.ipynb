```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
from matplotlib import pyplot as plt
```

```python
np.random.seed(42)
```

```python
non_stationary_data = np.cumsum(np.random.randn(100))
```

```python
non_stationary_data
```

```python
plt.plot(non_stationary_data)
```

```python
result = sm.tsa.adfuller(non_stationary_data)
```

### result
#### ADF statistics-> The ADF statistic measures how strongly your time series is affected by a unit root, which is an indicator of non-stationarity. A more negative ADF statistic suggests stronger evidence against the presence of a unit root and indicates stationarity if it's significantly negative.
#### P-value -> The p-value indicates the chance of getting the ADF statistic under the assumption that the series is non-stationary; a low p-value(below 0.05) rejects non-stationarity.
#### Number of Lags Used-> it indicates the number of lags included in the regression when calculating the ADF statistic.
#### Number of observations -> It denotes the sample size
#### Critical Values -> critical values are specific thresholds that are compared to the ADF statistic to determine whether the series is stationary or non-stationary.

```python
adf_statistic = result[0]
p_value = result[1]
```

```python
print(f'ADF Statistic: {adf_statistic}')
print(f'p-value: {p_value}')
```

```python
if p_value <= 0.05:
    print("Reject the null hypothesis (H0): The time series is stationary.")
else:
    print("Fail to reject the null hypothesis (H0): The time series is non-stationary.")
```
